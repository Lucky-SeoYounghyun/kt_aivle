## [실습2] 제대로 불러오기

- head(), tail() 메서드로 조회 결과를 볼 때, 세 번째 행부터 읽어와야 합니다.
- header=2를 지정하여 다시 불러와 seoul_people 데이터프레임으로 선언하세요.

```python
# 데이터 불러오기(세 번째 행부터)

seoul_people = pd.read_csv(path + '1.3 seoul_people_202407.csv', sep='\t', encoding='CP949', header = 2)

# 확인
seoul_people.head()
```

- seoul_people 데이터프레임이 '동별(2)', '남자', '여자', '소계.1' 열만 갖게 하세요.

```python
# '동별(2)','남자','여자','소계.1' 행만 갖게
# df = df[['열1', '열2', '열3']] 형태의 구문 사용

seoul_people = seoul_people[['동별(2)','남자','여자','소계.1']]
# 확인
seoul_people.head()
```

- 소계인 첫 번째 행은 의미가 없으니 제거하고, 인덱스를 초기화하세요.

```python
# 첫 번째 행 제거
# 참고: df.drop(인덱스) 형태 구문 사용 

seoul_people.drop(seoul_people.index[0], inplace=True)
# 확인
seoul_people.head()
```

```python
# 인덱스 초기화
# 참고: reset_index() 메서드 사용

seoul_people.reset_index()
# 확인
seoul_people
```

- 다음과 같이 seoul_people 데이터프레임의 열 이름을 변경하세요.
    - '동별(2)' --> '자치구', '소계.1' --> '합계'

```python
# 열 이름 변경: '동별(2)' --> '자치구', '소계.1' --> '합계'
# 참고: rename() 메서드 사용

seoul_people.rename(columns = {'동별(2)' : '자치구', '소계.1' : '합계'}, inplace = True)
# 확인

seoul_people
```

## [실습4] 시각화 분석

- 다양한 열를 기준으로 그래프를 그려보고 인사이트를 도출해보세요.

```python
# 상관계수

numeric_columns = seoul_people.select_dtypes(include=[np.number])
correlation = numeric_columns.corr()
print(correlation)
```

```python
# Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".8f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# barplot

plt.figure(figsize=(10, 6))
sns.barplot(x='자치구', y='남자', data=seoul_people)
plt.title('Bar Plot')
plt.xlabel('자치구')
plt.ylabel('총 이동인구')
plt.xticks(rotation=45)

plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

sns.boxplot(ax=axes[0], y=seoul_people['남자'])
axes[0].set_title('남자')

sns.boxplot(ax=axes[1], y=seoul_people['여자'])
axes[1].set_title('여자')

sns.boxplot(ax=axes[2], y=seoul_people['합계'])
axes[2].set_title('합계')

plt.show()
```

```python
# scatter
from scipy.stats import pearsonr

corr, p_value = pearsonr(seoul_people['남자'], seoul_people['여자'])
print(f"상관계수 : {corr}")
print(f"P-value : {p_value}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='남자', y='여자', data=seoul_people)
plt.title('Scatter')
plt.xlabel('남자')
plt.ylabel('여자')
plt.show()
```
