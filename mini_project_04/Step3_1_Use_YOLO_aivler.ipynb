```python
import os
import zipfile
import cv2
```

```python
def extract_zip_to_folder(zip_path, extract_folder_name):
    """
    ì£¼ì–´ì§„ ZIP íŒŒì¼ ê²½ë¡œë¥¼ '/content/image_data/' ê²½ë¡œì— ì••ì¶• í•´ì œí•˜ëŠ” í•¨ìˆ˜.

    Args:
    zip_path (str): ì••ì¶• í•´ì œí•  ZIP íŒŒì¼ì˜ ì „ì²´ ê²½ë¡œ.
    """
    # ê¸°ë³¸ ê²½ë¡œ ì„¤ì •
    zip_file_name = os.path.basename(zip_path)
    extract_folder = f'/content/image_data/{extract_folder_name}'

    # ì§€ì •ëœ ê²½ë¡œì— í´ë”ê°€ ì—†ì„ ë•Œ ìƒì„±
    if not os.path.exists(extract_folder):
        os.makedirs(extract_folder)
        print(f"{extract_folder} í´ë”ë¥¼ ìƒì„±í–ˆìŠµë‹ˆë‹¤.")

    # ZIP íŒŒì¼ ì••ì¶• í•´ì œ
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        for f in zip_ref.namelist():
            # ë””ë ‰í† ë¦¬ê°€ ì•„ë‹Œ ëª¨ë“  íŒŒì¼ì— ëŒ€í•´ ì²˜ë¦¬
            if not f.endswith('/'):
                # íŒŒì¼ì˜ ì „ì²´ ê²½ë¡œ ìƒì„±
                d_path = os.path.join(extract_folder, f)

                # íŒŒì¼ì˜ ë””ë ‰í† ë¦¬ ê²½ë¡œ ìƒì„± (í•˜ìœ„ ë””ë ‰í† ë¦¬ í¬í•¨)
                os.makedirs(os.path.dirname(d_path), exist_ok=True)

                # íŒŒì¼ ì¶”ì¶œ
                with zip_ref.open(f) as source, open(d_path, 'wb') as target:
                    target.write(source.read())

    print(f"{zip_path}ì˜ ì••ì¶•ì„ {extract_folder}ì— í•´ì œí–ˆìŠµë‹ˆë‹¤.")
```

```python
# import os
# import cv2

# def crop_images_to_center(src_folder, crop_size=(640, 640)):
#     """
#     ì£¼ì–´ì§„ í´ë”ì™€ ëª¨ë“  í•˜ìœ„ í´ë” ë‚´ì˜ .jpg ì´ë¯¸ì§€ë¥¼ ì¤‘ì‹¬ì„ ê¸°ì¤€ìœ¼ë¡œ ì£¼ì–´ì§„ í¬ê¸°ë¡œ í¬ë¡­í•˜ì—¬ ì›ë³¸ íŒŒì¼ì— ë®ì–´ì“°ëŠ” í•¨ìˆ˜.

#     Args:
#     src_folder (str): ì›ë³¸ ì´ë¯¸ì§€ í´ë” ê²½ë¡œ.
#     crop_size (tuple): í¬ë¡­í•  ì´ë¯¸ì§€ì˜ í¬ê¸° (ë„ˆë¹„, ë†’ì´).
#     """
#     # í´ë” ë‚´ ëª¨ë“  íŒŒì¼ íƒìƒ‰
#     for root, _, files in os.walk(src_folder):
#         for img_file in files:
#             if img_file.endswith('.jpg'):
#                 img_path = os.path.join(root, img_file)
#                 img = cv2.imread(img_path)

#                 if img is not None:
#                     h, w, _ = img.shape
#                     crop_w, crop_h = crop_size

#                     # ì¤‘ì‹¬ ì¢Œí‘œ ê³„ì‚°
#                     center_x, center_y = w // 2, h // 2

#                     # í¬ë¡­ ì˜ì—­ ê³„ì‚°
#                     x1 = max(center_x - crop_w // 2, 0)
#                     y1 = max(center_y - crop_h // 2, 0)
#                     x2 = min(center_x + crop_w // 2, w)
#                     y2 = min(center_y + crop_h // 2, h)

#                     # í¬ë¡­ëœ ì´ë¯¸ì§€ ìƒì„±
#                     cropped_img = img[y1:y2, x1:x2]

#                     # í¬ë¡­ëœ ì´ë¯¸ì§€ ì €ì¥ (ë®ì–´ì“°ê¸°)
#                     cv2.imwrite(img_path, cropped_img)

#     print(f"{src_folder} ë‚´ì˜ ëª¨ë“  ì´ë¯¸ì§€ë¥¼ {crop_size} í¬ê¸°ë¡œ í¬ë¡­í•˜ì—¬ ë®ì–´ì¼ìŠµë‹ˆë‹¤.")

import os
import cv2

def crop_and_resize_images(src_folder, crop_size=(480, 480), target_size=(640, 640)):
    """
    ì£¼ì–´ì§„ í´ë” ë‚´ ëª¨ë“  .jpg ì´ë¯¸ì§€ë¥¼ ì¤‘ì‹¬ì„ ê¸°ì¤€ìœ¼ë¡œ í¬ë¡­í•˜ê³ , ì§€ì •ëœ í¬ê¸°ë¡œ í™•ëŒ€í•˜ì—¬ ì›ë³¸ íŒŒì¼ì— ë®ì–´ì“°ëŠ” í•¨ìˆ˜.

    Args:
    src_folder (str): ì›ë³¸ ì´ë¯¸ì§€ í´ë” ê²½ë¡œ.
    crop_size (tuple): ì¤‘ì‹¬ì„ ê¸°ì¤€ìœ¼ë¡œ í¬ë¡­í•  ì´ë¯¸ì§€ì˜ í¬ê¸° (ë„ˆë¹„, ë†’ì´).
    target_size (tuple): í¬ë¡­ í›„ ë¦¬ì‚¬ì´ì¦ˆí•  ì´ë¯¸ì§€ì˜ ìµœì¢… í¬ê¸° (ë„ˆë¹„, ë†’ì´).
    """
    crop_w, crop_h = crop_size
    target_w, target_h = target_size

    # í´ë” ë‚´ ëª¨ë“  íŒŒì¼ íƒìƒ‰
    for root, _, files in os.walk(src_folder):
        for img_file in files:
            if img_file.endswith('.jpg'):
                img_path = os.path.join(root, img_file)
                img = cv2.imread(img_path)

                if img is not None:
                    h, w, _ = img.shape

                    # ì¤‘ì‹¬ ì¢Œí‘œ ê³„ì‚°
                    center_x, center_y = w // 2, h // 2

                    # í¬ë¡­ ì˜ì—­ ê³„ì‚°
                    x1 = max(center_x - crop_w // 2, 0)
                    y1 = max(center_y - crop_h // 2, 0)
                    x2 = min(center_x + crop_w // 2, w)
                    y2 = min(center_y + crop_h // 2, h)

                    # í¬ë¡­ëœ ì´ë¯¸ì§€ ìƒì„±
                    cropped_img = img[y1:y2, x1:x2]

                    # 480x480ìœ¼ë¡œ ì˜ë¦° ì´ë¯¸ì§€ë¥¼ 640x640ìœ¼ë¡œ ë¦¬ì‚¬ì´ì¦ˆ
                    resized_img = cv2.resize(cropped_img, (target_w, target_h))

                    # ë¦¬ì‚¬ì´ì¦ˆëœ ì´ë¯¸ì§€ ì €ì¥ (ë®ì–´ì“°ê¸°)
                    cv2.imwrite(img_path, resized_img)

    print(f"{src_folder} ë‚´ì˜ ëª¨ë“  ì´ë¯¸ì§€ë¥¼ {crop_size}ë¡œ í¬ë¡­í•˜ê³  {target_size}ë¡œ ë¦¬ì‚¬ì´ì¦ˆí•˜ì—¬ ë®ì–´ì¼ìŠµë‹ˆë‹¤.")
```

```python
import os

def adjust_labels_for_resize(label_path, scale_factor):
    """
    ë ˆì´ë¸” íŒŒì¼ì˜ ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œë¥¼ ì´ë¯¸ì§€ ë¦¬ì‚¬ì´ì¦ˆ ë¹„ìœ¨ì— ë§ì¶° ë³€ê²½í•˜ëŠ” í•¨ìˆ˜.
    ê°€ë¡œ ë°©í–¥ë§Œ ì¡°ì •í•˜ê³  ìœ íš¨ì„± ê²€ì‚¬ë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤.

    Args:
    label_path (str): ë ˆì´ë¸” íŒŒì¼ ê²½ë¡œ.
    scale_factor (float): ì´ë¯¸ì§€ í™•ëŒ€ ë¹„ìœ¨.
    """
    with open(label_path, 'r') as f:
        lines = f.readlines()

    new_lines = []
    for line in lines:
        class_id, x_center, y_center, width, height = map(float, line.strip().split())

        # ê°€ë¡œ ë°©í–¥ ë¹„ìœ¨ ì ìš©
        x_center -= 0.025
        width *= scale_factor

        # ìœ íš¨ì„± ê²€ì‚¬: 1ë³´ë‹¤ í¬ë©´ 1ë¡œ ì œí•œ
        x_center = min(x_center, 1.0)
        width = min(width, 1.0)

        new_lines.append(f"{int(class_id)} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")

    # ë ˆì´ë¸” íŒŒì¼ ë®ì–´ì“°ê¸°
    with open(label_path, 'w') as f:
        f.writelines(new_lines)

def adjust_all_labels_in_folder(src_folder, scale_factor=640/480):
    """
    ì£¼ì–´ì§„ í´ë” ë‚´ì˜ ëª¨ë“  .txt íŒŒì¼ì— ëŒ€í•´ ë ˆì´ë¸” ì¢Œí‘œë¥¼ ì¡°ì •í•˜ëŠ” í•¨ìˆ˜.

    Args:
    src_folder (str): ë ˆì´ë¸” íŒŒì¼ì´ ìˆëŠ” í´ë” ê²½ë¡œ.
    scale_factor (float): ì´ë¯¸ì§€ ê°€ë¡œ ë°©í–¥ í™•ëŒ€ ë¹„ìœ¨.
    """
    for root, _, files in os.walk(src_folder):
        for file in files:
            if file.endswith('.txt'):
                label_path = os.path.join(root, file)
                adjust_labels_for_resize(label_path, scale_factor)
```

#### 1) Label íŒŒì¼ì˜ í´ë˜ìŠ¤ ê°’ ë³€ê²½

```python
import random
import shutil
import os
import glob
```

```python
# ì›ë³¸ í´ë” ê²½ë¡œ
source_folders = ['/content/image_data/not_me_01', '/content/image_data/not_me_02']
# ìƒˆë¡­ê²Œ ê²°í•©í•  í´ë” ê²½ë¡œ
destination_folder = '/content/image_data/not_me'

# not_me í´ë” ìƒì„±
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# í´ë”ë¥¼ ìˆœíšŒí•˜ë©° not_me í´ë”ì— ë³‘í•©
for source_folder in source_folders:
    for root, dirs, files in os.walk(source_folder):
        for file in files:
            # images ë° labels í´ë” êµ¬ì¡° ìœ ì§€
            rel_path = os.path.relpath(root, source_folder)
            dest_path = os.path.join(destination_folder, rel_path)
            os.makedirs(dest_path, exist_ok=True)

            # .txt íŒŒì¼ ì²˜ë¦¬
            if file.endswith('.txt'):
                source_file = os.path.join(root, file)
                dest_file = os.path.join(dest_path, file)

                with open(source_file, 'r') as f:
                    lines = f.readlines()

                # ì¤‘ë³µ ë¼ë²¨ ì œê±° ë° ì²« ë²ˆì§¸ ìˆ«ìë¥¼ 1ë¡œ ë³€ê²½
                unique_lines = set()  # ì¤‘ë³µì„ ì œê±°í•˜ê¸° ìœ„í•œ ì§‘í•©
                with open(dest_file, 'w') as f:
                    for line in lines:
                        parts = line.split()
                        parts[0] = '1'  # ì²« ë²ˆì§¸ ìˆ«ìë¥¼ 1ë¡œ ë³€ê²½
                        new_line = " ".join(parts)

                        # ì¤‘ë³µë˜ì§€ ì•Šì€ ê²½ìš°ë§Œ ê¸°ë¡
                        if new_line not in unique_lines:
                            unique_lines.add(new_line)
                            f.write(new_line + '\n')
            else:
                # .txt ì™¸ì˜ íŒŒì¼ì€ ê·¸ëŒ€ë¡œ ë³µì‚¬
                shutil.copy2(os.path.join(root, file), dest_path)
```

```python
train_folder = '/content/image_data/not_me/test/images'

# train í´ë” ì•ˆì˜ íŒŒì¼ ìˆ˜ ê³„ì‚°
file_count = len([f for f in os.listdir(train_folder) if os.path.isfile(os.path.join(train_folder, f))])

print(f"'train' í´ë” ì•ˆì˜ íŒŒì¼ ìˆ˜: {file_count}")
```

#### 2) ëª¨ë¸ì´ ìš”í•˜ëŠ” êµ¬ì¡°ì˜ í´ë” ìƒì„±

```python
# ì›ë³¸ í´ë” ê²½ë¡œ
not_me_folder = '/content/image_data/not_me'
me_folder = '/content/image_data/me'
jihong_folder = '/content/image_data/jihong'
hyewon_folder = '/content/image_data/hyewon'
hyeongu_folder = '/content/image_data/hyeongu'
chaerim_folder = '/content/image_data/chaerim'

# Datasets í´ë” ê²½ë¡œ ë° í•˜ìœ„ í´ë” ìƒì„±
datasets_folder = '/content/Datasets'

# train, valid, test í´ë” ìƒì„± ë° í•˜ìœ„ images, labels í´ë” ìƒì„±
for split in ['train', 'valid', 'test']:
    os.makedirs(os.path.join(datasets_folder, split, 'images'), exist_ok=True)
    os.makedirs(os.path.join(datasets_folder, split, 'labels'), exist_ok=True)

# íŒŒì¼ ë³µì‚¬ í•¨ìˆ˜
def copy_files(src_folder, dest_folder_images, dest_folder_labels):
    for root, dirs, files in os.walk(src_folder):
        for file in files:
            src_path = os.path.join(root, file)
            if file.endswith('.jpg'):
                dest_path = os.path.join(dest_folder_images, file)
            elif file.endswith('.txt'):
                dest_path = os.path.join(dest_folder_labels, file)
            else:
                continue
            shutil.copy2(src_path, dest_path)

# ê° split í´ë”ì—ì„œ not_meì™€ me íŒŒì¼ì„ í†µí•©í•˜ì—¬ ë³µì‚¬
for split in ['train', 'valid', 'test']:
    # í†µí•© ëŒ€ìƒ í´ë” ê²½ë¡œ ì„¤ì •
    dest_images = os.path.join(datasets_folder, split, 'images')
    dest_labels = os.path.join(datasets_folder, split, 'labels')

    # not_me í´ë”ì—ì„œ ì´ë¯¸ì§€ ë° ë¼ë²¨ ë³µì‚¬
    not_me_src = os.path.join(not_me_folder, split)
    copy_files(not_me_src, dest_images, dest_labels)

    # me í´ë”ì—ì„œ ì´ë¯¸ì§€ ë° ë¼ë²¨ ë³µì‚¬
    me_src = os.path.join(me_folder, split)
    copy_files(me_src, dest_images, dest_labels)

    # jihong í´ë”ì—ì„œ ì´ë¯¸ì§€ ë° ë¼ë²¨ ë³µì‚¬
    jihong_src = os.path.join(jihong_folder, split)
    copy_files(jihong_src, dest_images, dest_labels)

    # me í´ë”ì—ì„œ ì´ë¯¸ì§€ ë° ë¼ë²¨ ë³µì‚¬
    hyewon_src = os.path.join(hyewon_folder, split)
    copy_files(hyewon_src, dest_images, dest_labels)

    # me í´ë”ì—ì„œ ì´ë¯¸ì§€ ë° ë¼ë²¨ ë³µì‚¬
    hyeongu_src = os.path.join(hyeongu_folder, split)
    copy_files(hyeongu_src, dest_images, dest_labels)

    # me í´ë”ì—ì„œ ì´ë¯¸ì§€ ë° ë¼ë²¨ ë³µì‚¬
    chaerim_src = os.path.join(chaerim_folder, split)
    copy_files(chaerim_src, dest_images, dest_labels)
```

```python
train_folder = '/content/Datasets/test/images'

# train í´ë” ì•ˆì˜ íŒŒì¼ ìˆ˜ ê³„ì‚°
file_count = len([f for f in os.listdir(train_folder) if os.path.isfile(os.path.join(train_folder, f))])

print(f"'test' í´ë” ì•ˆì˜ íŒŒì¼ ìˆ˜: {file_count}")
```
